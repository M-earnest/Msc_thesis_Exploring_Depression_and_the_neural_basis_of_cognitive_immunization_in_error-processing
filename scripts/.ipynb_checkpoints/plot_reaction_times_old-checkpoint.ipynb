```python
fig, ax = plt.subplots(figsize=(15,10))
loc = plticker.MultipleLocator(base=1.0) # this locator puts ticks at regular intervals
ax.xaxis.set_major_locator(loc)

ax.plot(rt_incongruent['mean rt'].to_numpy(), color='r')
ax.plot(rt_incongruent['mean rt'].to_numpy()- rt_incongruent['std'].to_numpy(), color='r', alpha=0.2)
ax.plot(rt_incongruent['mean rt'].to_numpy()+ rt_incongruent['std'].to_numpy(), color='r', alpha=0.2)

ax.fill_between(np.arange(len(rt_incongruent['mean rt'].to_numpy())),
                 rt_incongruent['mean rt'].to_numpy() - rt_incongruent['std'].to_numpy(),
                 rt_incongruent['mean rt'].to_numpy() + rt_incongruent['std'].to_numpy(), color='r', alpha=0.05)

ax.plot(rt_congruent['mean rt'].to_numpy())
ax.plot(rt_congruent['mean rt'].to_numpy()- rt_congruent['std'].to_numpy(), color='b', alpha=0.2)
ax.plot(rt_congruent['mean rt'].to_numpy()+ rt_congruent['std'].to_numpy(), color='b', alpha=0.2)

ax.fill_between(np.arange(len(rt_congruent['mean rt'].to_numpy())),
                 rt_congruent['mean rt'].to_numpy() - rt_congruent['std'].to_numpy(),
                 rt_congruent['mean rt'].to_numpy() + rt_congruent['std'].to_numpy(), color='b', alpha=0.05)

fig.canvas.draw()

labels = [item.get_text() for item in ax.get_xticklabels()]
labels[1] = 'Baseline'
labels[2] = 'Negative'
labels[3] = 'Positive'

ax.set_xticklabels(labels)

plt.ylim(0.2, 0.5)

incongruent_leg = mpatches.Patch(color='r', label='incongruent')
congruent_leg = mpatches.Patch(color='b', label='congruent')

plt.legend(handles=[incongruent_leg, congruent_leg])

plt.show()
```
