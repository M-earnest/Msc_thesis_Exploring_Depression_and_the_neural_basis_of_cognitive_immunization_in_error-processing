```python
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import matplotlib.patches as mpatches
import numpy as np
import os
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score #model evaluation
```

```python
rootdir = '/home/michael/data/derivatives/results/dataframes'
```

```python
rt_df = pd.DataFrame()

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        #print(file)
        #print os.path.join(subdir, file)
        filepath = subdir + os.sep + file
        if filepath.endswith(".tsv"):
            print (filepath)
            if filepath == '/home/michael/data/derivatives/results/dataframes/debugging_rt_data_sub_x.tsv':
                print('no, no')
            else:
                df_temp = pd.read_csv(filepath, sep='\t',  index_col=0) 
                rt_df = rt_df.append(df_temp)
                display(df_temp)
                
rt_df.sort_values(by=('subject'))
```

```python
list_subj_correct = []
list_num_correct = []
list_mean_rt_correct = []
list_sem_correct = []

list_subj_err = []
list_num_err = []
list_mean_rt_err = []
list_sem_err = []
# group by incongruent or congruent target in flanker-task

grouped = rt_df.groupby('subject')

# iterate over split dataframes
for name, group in grouped:
    #print(name)
    #display(group)
    #display(group['reaction'])
    rt_congruent_df = group
    
    
    incorrects = group.groupby('reaction')

    # iterate over split dataframes
    for name_incorrect, group_incorrect in incorrects:
        #print(name_incorrect)
        #display(group)
        if name_incorrect == 'correct':
            #print('number of correct answers subject ' + str(name) + ': ' + str(group_incorrect['reaction'].count()))
            ## add number of correct answers to df
            list_subj_correct.append(name)
            list_num_correct.append(group_incorrect['reaction'].count())
            list_mean_rt_correct.append(group_incorrect['rt'].mean())
            list_sem_correct.append(group_incorrect['rt'].sem())
            
        elif name_incorrect == 'incorrect':
            print('number of errors subject ' + str(name) + ': ' + str(group_incorrect['reaction'].count()))
            list_subj_err.append(name)
            list_num_err.append(group_incorrect['reaction'].count())
            list_mean_rt_err.append(group_incorrect['rt'].mean())
            list_sem_err.append(group_incorrect['rt'].sem())


            ## add number of incorrect answers to df
        else:
            print('nan')
            
            ## add number of incorrect answers to df
```

```python
subj_rt_df = pd.DataFrame({'subj':list_subj_correct,
                           'num_errors':list_num_err,
                           'num_correct':list_num_correct,
                           'mean_rt_err':list_mean_rt_err,
                           'mean_rt_correct':list_mean_rt_correct,
                           'sem_err': list_sem_err,
                           'sem_correct': list_sem_correct})
```
