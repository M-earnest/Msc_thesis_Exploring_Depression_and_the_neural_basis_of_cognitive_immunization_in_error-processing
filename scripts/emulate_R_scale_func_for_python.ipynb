```python
df_power = pd.read_csv('/home/michael/git/master_thesis/data/average_power_itc_by_condition_group__long_format.csv',
                         sep='\t')
df_power_plotting = pd.read_csv('/home/michael/git/master_thesis/data/average_power_itc_by_condition_group_long_format_plotting.csv',
                         sep='\t')
```

```python
bdi_z = (df_power['BDI_sum_score'] - df_power['BDI_sum_score'].mean()) / df_power['BDI_sum_score'].std()
des_z = (df_power['DES_sum_score'] - df_power['DES_sum_score'].mean()) / df_power['DES_sum_score'].std()
bdi_z_plotting = (df_power_plotting['BDI_sum_score'] - df_power_plotting['BDI_sum_score'].mean()) / df_power_plotting['BDI_sum_score'].std()
des_z_plotting = (df_power_plotting['DES_sum_score'] - df_power_plotting['DES_sum_score'].mean()) / df_power_plotting['DES_sum_score'].std()
```

```python
df_power['bdi_z'] = bdi_z
df_power['des_z'] = des_z
df_power_plotting['bdi_z'] = bdi_z_plotting
df_power_plotting['des_z'] = des_z_plotting
```
