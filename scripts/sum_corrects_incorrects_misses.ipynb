```python
list_subj = []
list_amount = []
list_mean_rt = []
list_sem = []

list_correct = []

list_congruent_incongruent = []
list_conditions = []
list_order = []

list_err_per_trials = []  # num_error/1248
list_err_per_trial_condition = []  # num_error/400
# group by incongruent or congruent target in flanker-task
temp_df_incorrect_congruent = pd.DataFrame(columns=('amount','condition'))
temp_df_correct_congruent = pd.DataFrame(columns=('amount','condition'))

temp_df_correct_incongruent = pd.DataFrame(columns=('amount','condition'))
temp_df_incorrect_incongruent = pd.DataFrame(columns=('amount','condition'))

temp_df_miss_incongruent = pd.DataFrame(columns=('amount','condition'))
temp_df_miss_congruent = pd.DataFrame(columns=('amount','condition'))


# group by incongruent or congruent target in flanker-task
congruent = df_temp.groupby('target')

# iterate over split dataframes
for name_congruent, group_congruent in congruent:
    print(name_congruent)
    #display(group_congruent)
    if name_congruent == 'congruent':
        incorrects = group_congruent.groupby('answer')
        for name_incorrect, group_incorrect in incorrects:
            #print(name_incorrect)
            #display(group_incorrect)
            if name_incorrect == 'correct':
                temp_df_correct_congruent = group_incorrect

                #print('correct congruent')

            # for incorrects
            elif name_incorrect == 'incorrect':
                temp_df_incorrect_congruent = group_incorrect              

                #print('incorrect congruent')
            elif name_incorrect == 'miss':
                temp_df_miss_congruent = group_incorrect                


    elif name_congruent == 'incongruent':

        incorrects = group_congruent.groupby('answer')

        # iterate over split dataframes
        for name_incorrect, group_incorrect in incorrects:

            if name_incorrect == 'correct':
                temp_df_correct_incongruent = group_incorrect

                #print('correct congruent')

            # for incorrects
            elif name_incorrect == 'incorrect':
                temp_df_incorrect_incongruent = group_incorrect             
                
                #print('incorrect congruent')
            elif name_incorrect == 'miss':
                temp_df_miss_incongruent = group_incorrect   

```
