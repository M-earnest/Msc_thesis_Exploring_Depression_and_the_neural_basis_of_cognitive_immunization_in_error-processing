```python
import pandas as pd
import numpy as np
```

```python
bdi = pd.read_csv('/home/michael/git/master_thesis/data/bdi_sum_std.tsv', sep='\t')
des = pd.read_csv('/home/michael/git/master_thesis/data/des_sum_std.tsv', sep='\t')
sps = pd.read_csv('/home/michael/git/master_thesis/data/sp_sum_std.tsv', sep='\t')
bdi_std = 4.672105673951675
des_std = 5.200427332868372
sps_std = 9.887943596695349
bdi_alpha = .80
des_alpha = .93
sps_alpha = .94
```

```python
list_upper_ci = []
list_lower_ci = []
list_sde = []

#sde = std * np.sqrt(1-a)
# ci = mx +/- sde * z
# kritischer z-wert (0.5 fehlerwahrscheinlichkeiz, zweiseitig) = +/- 1.960
for index in range(len(bdi)):
    print('subj :' + str(index))
    sde = bdi_std * np.sqrt(1-bdi_alpha)
    upper_ci = bdi['bdi_sum_score'][index] + sde * 1.960
    lower_ci = bdi['bdi_sum_score'][index] - sde * 1.960
    print(sde)
    print(bdi['bdi_sum_score'][index])
    print(upper_ci)
    print(lower_ci)
    list_upper_ci.append(upper_ci)
    list_lower_ci.append(lower_ci)
    list_sde.append(sde)
#bdi['sde'] = list_sde 
bdi['upper_ci'] = list_upper_ci
bdi['lower_ci'] = list_lower_ci
```
